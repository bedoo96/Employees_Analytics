```python
# Read a csv file and print its contents using pandas
import csv  
import pandas as pd

data = pd.read_excel('./Data/November-2025.xlsx', skiprows=1)
data.head()
```

```python
# print from the 6 column to the end
data.iloc[:, 33:]
#delete them from the dataframe
data = data.iloc[:, :33]    
data.head()
```

```python
#now replace the 01 , 02 by the day of the last month like 11/01/2025, take it from the system date make a loop through the first line and replace the values 01, 02 ... by the last month day

from datetime import datetime
now = datetime.now()
last_month = now.month - 1 if now.month > 1 else 12
year = now.year if now.month > 1 else now.year - 1
days_in_last_month = (datetime(year, last_month + 1, 1) - datetime(year, last_month, 1)).days
for i in range(6, 6 + days_in_last_month):
    day_str = f"{i - 5:02d}"
    new_date_str = f"{last_month:02d}/{day_str}/{year}"
    data.rename(columns={day_str: new_date_str}, inplace=True)
data.head()

```

```python
for col in data.columns[3:]:
    checkin_col = f"{col}_checkin"
    checkout_col = f"{col}_checkout"
    data[checkin_col] = data[col].apply(lambda x: x.split('-')[0] if isinstance(x, str) and '-' in x else None)
    data[checkout_col] = data[col].apply(lambda x: x.split('-')[1] if isinstance(x, str) and '-' in x else None)
    data.drop(col, axis=1, inplace=True)  # Remove original column
```

```python
data.head(20)
```

```python

import calendar

# Filter columns that look like dates (contain '/')
date_cols = [col for col in data.columns[4:] if '/' in col]

for col in date_cols:
    # Remove any suffix after '_' (e.g., "_checkin" or "_checkout")
    base_date = col.split('_')[0]  # "11/02/2025"
    month, day, year = map(int, base_date.split('/'))
    
    weekday = calendar.weekday(year, month, day)
    if weekday in (5, 6):  # Saturday or Sunday
        checkin_col = f"{base_date}_checkin"
        checkout_col = f"{base_date}_checkout"
        # Drop only if columns exist
        for c in [checkin_col, checkout_col]:
            if c in data.columns:
                data.drop(c, axis=1, inplace=True)

# Save processed file
data.head(20).to_excel('./Data/November-2025-Processed.xlsx', index=False)
data.head(20)

```

```python

from datetime import datetime

# Get all check-in columns dynamically
checkin_cols = [col for col in data.columns if col.endswith('_checkin')]

def calculate_total_hours(row):
    total_hours = 0
    for checkin_col in checkin_cols:
        checkout_col = checkin_col.replace('_checkin', '_checkout')
        checkin = row.get(checkin_col)
        checkout = row.get(checkout_col)
        if checkin and checkout:
            try:
                checkin_time = datetime.strptime(checkin, '%H:%M')
                checkout_time = datetime.strptime(checkout, '%H:%M')
                hours_worked = (checkout_time - checkin_time).seconds / 3600
                total_hours += hours_worked
            except ValueError:
                continue
    return round(total_hours, 2)

data['Total_Hours'] = data.apply(calculate_total_hours, axis=1)
data.head(20)
```
